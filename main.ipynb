```python
!pip install rarfile

from google.colab import drive
drive.mount('/content/drive')

import os
import rarfile

# Set the directory path containing the subfolders with the splitted RAR files in Google Drive
dir_path = "/content/drive/MyDrive/path/to/subfolders"

# Set the path of the directory where the extracted files will be saved in Google Drive
extract_dir = "/content/drive/MyDrive/path/to/extract/the/files/to"

# Set the password for the password-protected RAR files
password = "your_password_here"

# Keep running the loop until all RAR files have been extracted
while True:
    # Set a flag to keep track of whether any RAR files were extracted during the current loop
    extracted = False

    # Loop through all subfolders in the directory
    for subfolder in os.listdir(dir_path):
        subfolder_path = os.path.join(dir_path, subfolder)
        if os.path.isdir(subfolder_path):
            # Loop through all RAR files in the subfolder
            for file in os.listdir(subfolder_path):
                if file.endswith(".part1.rar") or file.endswith(".part01.rar"):
                    # Create a folder with the same name as the subfolder
                    extract_folder = os.path.join(extract_dir, subfolder)
                    os.makedirs(extract_folder, exist_ok=True)

                    # Join the splitted RAR files to get the full file path
                    file_path = os.path.join(subfolder_path, file)

                    # Open the RAR file using rarfile module
                    with rarfile.RarFile(file_path) as rar:
                        # Check if the RAR file is password protected
                        if rar.needs_password():
                            rar.setpassword(password)
                        # Extract all files to the extract folder
                        rar.extractall(path=extract_folder)

                    print(f"Extraction completed successfully for {file} in {subfolder}!")
                    extracted = True

    # If no RAR files were extracted during the current loop, exit the while loop
    if not extracted:
        print("All RAR files have been extracted successfully!")
        break
```
